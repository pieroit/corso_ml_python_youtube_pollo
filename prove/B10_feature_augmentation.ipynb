Le feature possono essere "aumentate", combinate tra di loro per aiutare il modello a trovare più facilmente delle relaioni utili alla predizione.
- vediamo un tipo di task di classificazione non lineare, cioè un modello che deve separare in 2 parti una nuvola di punti che non sono separabili da una linea
- partiao dal dataset, che generiamo sinteticamente


```python
import seaborn as sns
import numpy as np

from sklearn.datasets import make_circles
from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_circles(n_samples=1000, noise=0.2, factor=0.2)
#X, y
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
model.score(X,y)
```

```python
p = model.predict(X)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=p)
```

```python
allX = np.random.random( (10000, 2) ) * 3 - 1.5
allX
```

```python
allp = model.predict(allX)
```

```python
sns.scatterplot(x=allX[:,0], y=allX[:,1], hue=allp)
```

ottengo nuove feature moltiplicando tra loro le feature esistenti o applicando funzioni non lineari (potenza, log, trigonometria)

```python
augmenter = PolynomialFeatures(2)
augmenter.fit(X)
X_plus = augmenter.fit_transform(X)
```

```python
augmenter.get_feature_names()
```

```python
model_plus = LogisticRegression()
model_plus.fit(X_plus, y)
model_plus.score(X_plus, y)
```

```python
p_plus = model_plus.predict(X_plus)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=p_plus)
```
