```python
from sklearn.datasets import load_digits
from sklearn.naive_bayes import MultinomialNB
#!pip install yellowbrick
from yellowbrick.model_selection import learning_curve

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.dummy import DummyClassifier
```

```python
X, y = load_digits(return_X_y=True)

model = MultinomialNB()
```

- con la curva di apprendimento posso vedere come variano le prestazioni su train e test al crescere dei dati
- l'obiettivo è capire qual'è la giusta quantità di dati da utilizzare nel training
 - se train e test convergono, la quantità di dati è sufficiente
 - se il divario è alto (overfitting), ho beneficio dall'aggiungere dati
 - se il train resta basso, ho problemi con l'apprendimento a prescindere dai dati
- cv mi permette di controllare le fold della crossvalidation

```python
learning_curve(model, X, y, cv=3, scoring='accuracy')
```

```python
models = [
    MultinomialNB(),
    DecisionTreeClassifier(),
    KNeighborsClassifier(),
    DummyClassifier()
]
```

```python
for m in models:
    learning_curve(m, X, y)
```
