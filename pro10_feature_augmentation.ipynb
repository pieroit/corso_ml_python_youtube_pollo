```python
import seaborn as sns
import numpy as np

from sklearn.datasets import make_circles
from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_circles(n_samples=1000, noise=0.2, factor=0.2)
```

```python
x0 = X[:,0]
x1 = X[:,1]
```

```python
sns.scatterplot(x=x0, y=x1, hue=y)
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
p = model.predict(X)
```

```python
sns.scatterplot(x=x0, y=x1, hue=p)
```

```python
augmenter = PolynomialFeatures(2)
```

```python
augmenter.fit(X)
X_plus = augmenter.transform(X)
```

```python
augmenter.get_feature_names()
```

```python
model = LogisticRegression()
model.fit(X_plus, y)
```

```python
model.score(X_plus, y)
```

```python
p_plus = model.predict(X_plus)
```

```python
sns.scatterplot(x=x0, y=x1, hue=p_plus)
```
